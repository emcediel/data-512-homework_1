```python
# These are standard python modules
import json

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

We read the json files created in the clean_data notebook

```python
#We read the json files 
f = open('../data_clean/dino_monthly_mobile_201507-202210.json')
monthly_mobile = json.load(f)
f.close()
f = open('../data_clean/dino_monthly_desktop_201507-202210.json')
monthly_desktop = json.load(f)
f.close()
f = open('../data_clean/dino_monthly_cumulative_201507-202210.json')
monthly_cummulative = json.load(f)
f.close()
```

We create a function that converts the json files into dataframes

```python
# Function: Converts wikimedia json to a DataFrame
# Input: clean JSON file from Wikimedia
# Output" DataFrame 
def json_to_df(json):
    df = pd.DataFrame()
    for dino, months in json.items():
        df_temp = pd.DataFrame.from_dict(months)
        df = pd.concat([df,df_temp])
    #We correct the timestamp format
    df['timestamp'] = df['timestamp'].apply(lambda x: x[:6])
    df['date'] = pd.to_datetime(df['timestamp'], format = '%Y%m')
    return df
```

## Top 10 Peak Page Views ##  
 The second graph should contain time series for the top 10 article pages by largest (peak) page views over the entire time by access type. You first find the month for each article that contains the highest (peak) page views, and then order the articles by these peak values. Your graph should contain the top 10 for desktop and top 10 for mobile access (20 lines).

We find the top 10 highest mobile page views using groupby's. We create additional helper columns that will come handy later

```python
#Top 10 highest mobile page views  
top_10_views_mobile = mobile_df.groupby('article').max('views').sort_values('views', ascending = False)[:10].reset_index()
top_10_views_mobile.rename({'views':'peak_views'},axis=1, inplace=True)
top_10_views_mobile['top'] = top_10_views_mobile.index +1
top_10_views_mobile['peak page views'] = top_10_views_mobile['top'].apply(lambda x:'mobile_top_'+str(x) )
top_10_views_mobile
```

We do the dame for desktop

```python
#Top 10 highest desktop page views  
top_10_views_desktop = desktop_df.groupby('article').max('views').sort_values('views', ascending = False)[:10].reset_index()
top_10_views_desktop.rename({'views':'peak_views'},axis=1, inplace=True)
top_10_views_desktop['top'] = top_10_views_desktop.index +1
top_10_views_desktop['peak page views'] = top_10_views_desktop['top'].apply(lambda x:'desktop_top_'+str(x) )
top_10_views_desktop
```

In order to plot in a organized way, we do some additional modifications to the DataFrame, this specifically acommodates the DataFrame to what Seaborn expects to read. 

We make an inner join between the original DataFrame and the groupby DataFrame to obtain only the articles of interest. We sort them in order to have the highest peaks first

We call additional functions for plot formatting and then save the line plot in the results folder.

```python
#We adjust a dataframe for plotting and plot the result
peak_views_desktop = desktop_df.merge(top_10_views_desktop, on='article', how='inner')
peak_views_mobile = mobile_df.merge(top_10_views_mobile, on='article', how='inner')
peak_views = pd.concat([peak_views_desktop, peak_views_mobile]).reset_index()
peak_views.sort_values(by=['top'], inplace=True)
fig, axes = plt.subplots(figsize=(20,12))

#Plot formatting
years = mdates.YearLocator()   # every year
months = mdates.MonthLocator()  # every month
years_fmt = mdates.DateFormatter('%Y-%m')
axes.xaxis.set_major_locator(months)
axes.xaxis.set_major_formatter(years_fmt)
axes.xaxis.set_minor_locator(months)
axes.tick_params(axis='x', labelrotation=90)

#Plotting
sns.lineplot(data=peak_views, x='date',y='views',hue='peak page views').set(title='Wikipedia top 10 peak page dinosaur article views for mobile and desktop')
plt.savefig('../results/top_10_peak_page_views.png')
```

## Fewest Months of Data ##  
The third graph should show pages that have the fewest months of available data. These will all be relatively short time series, some may only have one month of data. Your graph should show the 10 articles with the fewest months of data for desktop access and the 10 articles with the fewest months of data for mobile access.


We find the top 10 articles with fewest months of available data using groupby's. We create additional helper columns that will come handy later

```python
#Fewest months of mobile data
fewest_months_mobile = mobile_df[['article','views']].groupby('article').count().sort_values('views')[:10].reset_index()
fewest_months_mobile.rename({'views':'months_of_data'},axis=1, inplace=True)
fewest_months_mobile['top'] = fewest_months_mobile.index +1
fewest_months_mobile['fewest months of data'] = fewest_months_mobile['top'].apply(lambda x:'mobile_top_'+str(x) )
fewest_months_mobile
```

We do the same for desktop data

```python
#Fewest months of desktop data
fewest_months_desktop = desktop_df[['article','views']].groupby('article').count().sort_values('views')[:10].reset_index()
fewest_months_desktop.rename({'views':'months_of_data'},axis=1, inplace=True)
fewest_months_desktop['top'] = fewest_months_desktop.index +1
fewest_months_desktop['fewest months of data'] = fewest_months_desktop['top'].apply(lambda x:'desktop_top_'+str(x) )
fewest_months_desktop
```

In order to plot in a organized way, we do some additional modifications to the DataFrame, this specifically acommodates the DataFrame to what Seaborn expects to read.

We make an inner join between the original DataFrame and the groupby DataFrame to obtain only the articles of interest. We sort them in order to have the articles with lowest amount of data first. 

We call additional functions for plot formatting and then save the line plot in the results folder.

```python
#We adjust a dataframe for plotting and plot the result
fewest_months_desktop = desktop_df.merge(fewest_months_desktop, on='article', how='inner')
fewest_months_mobile = mobile_df.merge(fewest_months_mobile, on='article', how='inner')
fewest_months = pd.concat([fewest_months_desktop, fewest_months_mobile]).reset_index()
fewest_months.sort_values(by=['top'], inplace=True)

#Plot formatting
fig, axes = plt.subplots(figsize=(20,12))
years = mdates.YearLocator()   # every year
months = mdates.MonthLocator()  # every month
years_fmt = mdates.DateFormatter('%Y-%m')
axes.xaxis.set_major_locator(months)
axes.xaxis.set_major_formatter(years_fmt)
axes.xaxis.set_minor_locator(months)
axes.tick_params(axis='x', labelrotation=90)

#Plotting
sns.lineplot(data=fewest_months, x='date',y='views',hue='fewest months of data').set(title='Wikipedia top 10 dinosaur article views with fewest months of available data views for mobile and desktop')
plt.savefig('../results/fewest_months_of_data.png')
```
